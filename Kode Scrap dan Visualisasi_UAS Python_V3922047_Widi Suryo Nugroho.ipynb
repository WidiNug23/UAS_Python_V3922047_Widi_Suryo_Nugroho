```python
import csv
import requests
from bs4 import BeautifulSoup as bs

# Membuat file CSV untuk menyimpan data
csv_file = open('bukalapak_scraping.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Halaman', 'Title', 'Price'])

# URL halaman yang akan di-scrape
base_url = "https://www.bukalapak.com/u/barokah_kamera"

# Mengirim permintaan GET dan melakukan scraping hingga halaman 5
for page in range(1, 11):
    url = base_url + "?page=" + str(page)
    response = requests.get(url)
    soup = bs(response.content, 'html.parser')

    # Mencari elemen-elemen yang mengandung judul dan harga item
    items = soup.find_all('div', class_='c-product-card-description')

    # Menyimpan judul dan harga item ke dalam file CSV
    for item in items:
        title = item.find('a', class_='c-product-card__name').text.strip()
        price = item.find('span', class_='c-product-price-normal').text.strip()
        csv_writer.writerow([page, title, price])

# Menutup file CSV
csv_file.close()

print("Scraping selesai. Data disimpan dalam file 'V3922047_UASPython.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Memebaca file csv yang akan divisualisasikan datanya
data = pd.read_csv("V3922047_UASPython.csv")

# Memilih data hanya untuk halaman 1
data_page1 = data[data['Halaman'] == 1]

# Mengambil data Title dan data Price dari file
plt.bar(data_page1['Title'], data_page1['Price'])

# Memebuat judul untuk chart
plt.title("Data Camera Bar Chart (Page 1)")  

# Membuat label x dan label y
plt.xlabel('Camera')
plt.ylabel('Price')

# Menampilkan data visualisasi
plt.show()

```
